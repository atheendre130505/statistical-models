```python
import numpy as np
import matplotlib.pyplot as plt

def sierpinski_triangle(points, depth):
    if depth == 0:
        plt.fill(*zip(*points), 'k')
    else:
        mid1 = (points[0] + points[1]) / 2
        mid2 = (points[1] + points[2]) / 2
        mid3 = (points[2] + points[0]) / 2

        sierpinski_triangle([points[0], mid1, mid3], depth - 1)
        sierpinski_triangle([mid1, points[1], mid2], depth - 1)
        sierpinski_triangle([mid3, mid2, points[2]], depth - 1)

def draw_sierpinski_triangle(order=5):
    plt.figure(figsize=(6, 6))
    sierpinski_triangle(np.array([[0, 0], [1, 0], [0.5, np.sqrt(3)/2]]), order)
    plt.axis('off')
    plt.show()

def sierpinski_square(x, y, size, depth):
    if depth == 0:
        plt.fill([x, x + size, x + size, x], [y, y, y - size, y - size], 'k')
    else:
        new_size = size / 3
        for i in range(3):
            for j in range(3):
                if (i, j) != (1, 1):
                    sierpinski_square(x + i * new_size, y - j * new_size, new_size, depth - 1)

def draw_sierpinski_square(order=4):
    plt.figure(figsize=(6, 6))
    sierpinski_square(0, 0, 1, order)
    plt.axis('off')
    plt.show()

draw_sierpinski_triangle(7)
draw_sierpinski_square(4)
```
