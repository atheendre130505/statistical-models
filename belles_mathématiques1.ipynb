```python
import numpy as np
import matplotlib.pyplot as plt

def dragon_curve(n):
    x, y = [0], [0]

    directions = [1]
    for _ in range(n):
        new_directions = directions + [1] + [-d for d in reversed(directions)]
        directions = new_directions

    angle = np.pi / 2
    x.append(1)
    y.append(0)

    for turn in directions:
        dx, dy = x[-1] - x[-2], y[-1] - y[-2]
        new_dx = dx * np.cos(angle * turn) - dy * np.sin(angle * turn)
        new_dy = dx * np.sin(angle * turn) + dy * np.cos(angle * turn)
        x.append(x[-1] + new_dx)
        y.append(y[-1] + new_dy)

    plt.figure(figsize=(8, 8))
    plt.plot(x, y, 'k', lw=1)
    plt.axis("equal")
    plt.axis("off")
    plt.show()

dragon_curve(4)
#notes: start with a line, now at the end point, take whatever existed before and place it perpendicular to your end point, now u will have a new end point
```
